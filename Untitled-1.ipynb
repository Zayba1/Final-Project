#### Data Preprocessing

```python
# Handle missing values
df1.dropna(subset=['sideEffectsReview', 'sideEffects'], inplace=True)

# Text preprocessing function
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

# Apply preprocessing
df1['processed_review'] = df1['sideEffectsReview'].apply(preprocess_text)
```

```python
# Tokenization and padding
MAX_WORDS = 5000
MAX_LENGTH = 200

tokenizer = Tokenizer(num_words=MAX_WORDS)
tokenizer.fit_on_texts(df1['processed_review'])
X_sequences = tokenizer.texts_to_sequences(df1['processed_review'])
X_padded = pad_sequences(X_sequences, maxlen=MAX_LENGTH)

# Encode labels
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df1['sideEffects'])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_padded, y, 
    test_size=0.2, 
    random_state=42,
    stratify=y
)

# Print shapes of the resulting arrays
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
print("Number of classes:", len(label_encoder.classes_))
print("\nLabel mapping:")
for i, label in enumerate(label_encoder.classes_):
    print(f"{label}: {i}")
```

# Model Building

### We'll implement two different deep learning architectures:
#### 1. Bidirectional LSTM
#### 2. CNN

```python
def build_lstm_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Embedding(MAX_WORDS, 128, input_length=MAX_LENGTH),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(len(label_encoder.classes_), activation='softmax')
    ])
    return model

def build_cnn_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Embedding(MAX_WORDS, 128, input_length=MAX_LENGTH),
        tf.keras.layers.Conv1D(128, 5, activation='relu'),
        tf.keras.layers.GlobalMaxPooling1D(),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(len(label_encoder.classes_), activation='softmax')
    ])
    return model
```

# Model Training and Evaluation

```python
# Initialize models
models = {
    'LSTM': build_lstm_model(),
    'CNN': build_cnn_model()
}

# Train and evaluate models
results = {}
for name, model in models.items():
    print(f"\nTraining {name} model...")
    model.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])
    
    history = model.fit(
        X_train, y_train,
        epochs=10,
        validation_data=(X_test, y_test),
        batch_size=32
    )
    
    results[name] = {
        'history': history.history,
        'evaluation': model.evaluate(X_test, y_test)
    }
```

# Results Visualization and Analysis

```python
# Plot training history
plt.figure(figsize=(12, 5))
for name, result in results.items():
    plt.plot(result['history']['accuracy'], label=f'{name} Training Accuracy')
    plt.plot(result['history']['val_accuracy'], label=f'{name} Validation Accuracy')
plt.title('Model Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Print final results
print("Final Model Performance:")
for name, result in results.items():
    print(f"\n{name} Model:")
    print(f"Test Loss: {result['evaluation'][0]:.4f}")
    print(f"Test Accuracy: {result['evaluation'][1]:.4f}")
```

## Comparative Analysis

Both LSTM and CNN models were trained for 10 epochs to classify side effects from drug reviews. Here's a detailed analysis:

### CNN Model Performance
- Final Test Accuracy: 55.35%
- Test Loss: 1.7093
- Training time: ~10 seconds per epoch
- Showed more stable validation metrics
- Better overall performance than LSTM

### LSTM Model Performance
- Final Test Accuracy: 52.88%
- Test Loss: 2.2965

- Higher computational cost
- More prone to overfitting

## Key Observations

1. Overfitting Issues:
   - Both models show clear signs of overfitting, with training accuracy reaching >90% while validation accuracy remains around 55%
   - LSTM showed more severe overfitting (training accuracy: 90.72% vs validation: 52.88%)
   - CNN's overfitting was also significant (training accuracy: 97.95% vs validation: 55.35%)

2. Model Efficiency:
   - CNN was significantly more efficient, training about 5x faster than LSTM
   - CNN achieved slightly better test accuracy with less computational resources

3. Learning Patterns:
   - Both models showed rapid initial improvement in early epochs
   - Performance plateaued on validation data after epoch 4-5
   - Validation loss increased in later epochs while training loss continued to decrease

## Recommendations

1. To address overfitting:
   - Implement stronger regularization (e.g., increased dropout, L2 regularization)
   - Reduce model complexity
   - Add early stopping with a shorter patience period
   - Consider using data augmentation techniques

2. For model improvement:
   - Experiment with hybrid architectures combining CNN and LSTM
   - Try different preprocessing techniques for the text data
   - Consider using pre-trained embeddings like Word2Vec or GloVe
   - Implement k-fold cross-validation for more robust evaluation

3. For practical deployment:
   - The CNN model is recommended due to better efficiency and slightly higher accuracy
   - Current performance (~55% accuracy) suggests room for improvement before production use
   - Consider ensemble methods to improve prediction reliability
